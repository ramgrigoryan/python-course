```python
bookstoreUrl = "https://books.toscrape.com/"
```

```python
bookstore_page = requests.get(bookstoreUrl)
```

```python
bookstore_soup = bs4.BeautifulSoup(bookstore_page.text,'lxml')
```

```python
books = bookstore_soup.select('.product_pod')
```

```python
books
```

```python
def filterBook(book):
    return book.select('.star-rating')[0]['class'][1]=='Two'
```

```python
twoStarBooks = list(filter(filterBook,books))
```

```python
twoStarBooks
```

```python
for book in twoStarBooks:
    print(book.select('h3')[0].text)
```

```python
def get2StarBooksTitles():
    twoStarBooks = []
    baseUrl = "https://books.toscrape.com/catalogue/page-{}.html"
    page = 1
    while page<=50:
        bookstore_page = requests.get(baseUrl.format(page))
        bookstore_soup = bs4.BeautifulSoup(bookstore_page.text,'lxml')
        books = bookstore_soup.select('.product_pod')
        twoStarBookBlocks = list(filter(filterBook,books))
        for book in twoStarBookBlocks:
            twoStarBooks.append(book.select('h3 a')[0]['title'])
        page +=1
    return twoStarBooks
```

```python
allTwoStarBooks = get2StarBooksTitles()
```
